```python
def programa_beneficio():
  identificación=int(input("ingresar identificación: "))
  nombres=input("ingresar los nombres: ")
  apellidos=input("ingresar los apellidos: ")
  direccion=input("ingresar Dirección de residencia: ")
  telefono=int(input("ingresar numero telefonico: "))
  edad=int(input("ingresar edad: "))
  estado_civil=int(input("ingresar numero segun su estado civil: 1.Casado 2.Soltero 3.Union Libre 4.Viudo "))
  n_hijos=int(input("ingresar numero de hijos: "))
  estatura=int(input("ingresar estatura en cm: "))
  f_contratación=input("ingresar fecha de contratacion dd/mm/aaaa: ")
  s_basico=int(input("ingresar salario basico: "))
  d_laborados=int(input("ingresar numero de días laborados: "))
  print("\n¡DATOS DEL EMPLEADO!")
  print("IDENTIFICACIÓN: ",identificación)
  print("NOMBRE: ",nombres,apellidos)
  print("DIRECCIÓN: ",direccion)
  print("TELEFONO: ",telefono)
  print("EDAD: ",edad)
  print("ESTADO CIVÍL: ",estado_civil)
  print("NÚMERO DE HIJOS: ",n_hijos)
  print("ESTATURA: ",estatura)
  print("FECHA DE CONTRATACIÓN: ",f_contratación)
  print("SALARIO BASICO: ",s_basico)
  print("DÍAS LABORADOS: ",d_laborados) 

  print("\nBENEFICIOS CORPORATIVOS") 

  print("\nCOMISIÓN ANTIGUEDAD ") 
  if edad>55:
    comision=int((s_basico*0.05))
    print("¡Beneficio de comision de = ${} ".format(comision))
  else:
    comision=0
    print("¡Beneficio de comision NO otorgado!")

  print("\nBENEFICIO FAMILIAR")
  if estado_civil==1:
    if n_hijos>0:
      print("¡Beneficio de Viaje en Diciembre otorgado!")
    else:
      print("¡Beneficio de Viaje en Diciembre NO otorgado!")  
  else:
    print("¡Beneficio de Viaje en Diciembre NO otorgado!") 

  print("\nCOMISIÓN NIVEL SALARIAL")
  if s_basico>=1000000 and s_basico<=1500000:
    comision2=int((s_basico*0.02))
    print("¡Beneficio de comision de = ${}".format(comision2))
  elif s_basico>1500000 and s_basico<=2000000:
    comision2=int((s_basico*0.05))
    print("¡Beneficio de comision de = ${}".format(comision2)) 
  else:
    print("¡Beneficio de comision NO otorgado!")
    comision2=0 

  print("\nBENEFICIO ALIMENTACIÓN")
  if d_laborados>20:
    if s_basico<1000000:
      print("¡Beneficio de Bono de alimentación otorgado!")
    else:
      print("¡Beneficio de Bono de alimentación NO otorgado!")  
  else:
    print("¡Beneficio de Bono de alimentación NO otorgado!")  

  print("\nTOTAL SALARIO MAS COMISIONES")
  c_total=comision+comision2
  s_total=c_total+s_basico 
  print("¡Beneficio de comisiones es de = ${} para un salario total de ${} ".format(c_total,s_total)) 

def tabla_multiplicar():
  numero = int(input("número al cual generar tabla: "))
  for i in range (11):
    print(str(numero)+" x "+str(i)+" = "+str(i*numero))

def bonificación_4(ini):
  while ini=="si" or ini=="yes":
    salario=int(input("\nIngrese salario: "))
    if salario>655000:
      print("\n¡Beneficio de comision NO otorgado!")
    
    else:
      bonificacion=int((salario*0.04))
      t_salario=bonificacion+salario
      print("\n¡Beneficio de comision de = ${} y Su sueldo total es de: ${} ".format(bonificacion,t_salario))
    ini=input("\nDesea iniciar la ejecución del programa BONIFICICACIÓN una vez mas: ").lower()

def registro_notas(ini):
  print("\n¡SISTEMA DE REGISTRO DE NOTAS!")
  s_prom=0
  n_alumnos=0
  while ini=="si" or ini=="yes":
    print("\nRegitro para el alumno n.{}".format(n_alumnos+1))
    s_nota=0
    for i in range(5):
      nota=int(input("Ingresar "+str(i+1)+"° nota: "))
      s_nota=s_nota+nota
    prom=(s_nota/5)
    n_alumnos=n_alumnos+1
    s_prom=s_prom+prom
    print("El promedio del estudiante {} es: {}".format(n_alumnos,prom))
    ini=input("\nDesea agregar un nuevo alumno (si(yes)-no): ").lower()
  p_alumnos=s_prom/n_alumnos
  print("\nEl promedio para los {} alumnos registrados es de: {}".format(n_alumnos,p_alumnos))

def pares_100(i,n):
  for i in range (0,101,2):
    sum=i+i    
    n=n+1
  print("Los numeros pares son: ",(n-1))
  print("La suma de los numeros pares es:",sum)            

def años_de_vida():
  print("\n¡AÑOS DE VIDA!")
  edad=int(input("Ingrese edad en este año(2023): "))
  a_nacimiento=2023-edad
  for i in range (1,edad):
    a_nacimiento=a_nacimiento+1
    print("Para el año {} cumplio {} años.".format(a_nacimiento,i))

def p_inicio(ini):
  while ini=="si" or ini=="yes":
    print("\n¡PROGRAMAS DE GESTION!")
    est = int(input("\nQue programa desea ejecutar (digite número 1-6):\n \n1.Beneficio\n2.Tabla de Multiplicar\n3.Bonificación 4%\n4.Registro notas\n5.Pares de 100\n6.Años\n7.Salir\n\n"))
    if est== 1:
      programa_beneficio()
    elif est==2:
      tabla_multiplicar()
    elif est==3:
      bonificación_4("yes")
    elif est==4:
      registro_notas("si")
    elif est==5:
      pares_100(0,0)
    elif est==6:
      años_de_vida()
    elif est==7:
      break
    else:
      print("\n¡Error! ¡Ingresar numero Valido!")

p_inicio("si")









```
